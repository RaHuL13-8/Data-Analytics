```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
'''Read about the movielens dataset and write down a summary of metadata.'''


'''- MovieLens is a movie rec service

- Dataset has 5 star ratings and free text activity

- 100,836 ratings (0.5-5 w/ 0.5 increments)

- 9742 movies

- 3683 tags

- Tags are user generated metadata about movies.Each contains single word or a short phrase.

- Users represented by ID (consistent b/w rating and tags)

- All selected users had rated atleast 20 movies.

- Movie ID consistent accross ratings, tags, movies and links

- Data is contained in files 'links.csv','movies.csv','ratings.csv','tags.csv'

- Genres pipe separated list'''
```

```python
df=pd.read_csv("ratings.csv")
movies=pd.read_csv("movies.csv")
tags=pd.read_csv("tags.csv")
links=pd.read_csv("links.csv")
df_1=pd.merge(df,movies,on='movieId')
```

```python
''' Read the “ratings.csv” file and create a pivot table with index=‘userId’, columns=‘movieId’,
values = “rating.'''
pivot = df.pivot_table(index=['userId'], columns=['movieId'], values=['rating'])
pivot=pivot.fillna(0)
pivot
```

```python
''' sklearn.metrics.pairwise_distances can be used to compute distance between all pairs of 
users. pairwise_distances() takes a metric parameter for what distance measure to use. Use 
cosine similarity for finding similarity among users. Use the following packages.'''
distance=pairwise_distances(pivot,metric="cosine")
x= pd.DataFrame(np.squeeze(distance))
x
```

```python
'''Find the 5 most similar user for user with user Id 25'''
user_10=df['userId']==10
allusers=pivot.values
user_10=allusers[0]
den1=np.sqrt(sum([np.square(x) for x in user_10]))
cosinesim=[(10,1)]
i=1
for user in allusers[1:]:
    numer=[x*y for x,y in zip(user_10,user)]
    den2=np.sqrt(sum([np.square(x) for x in user]))
    costheta = sum(numer)/(den1*den2)
    cosinesim.append((pivot.index[i],costheta))
    i=i+1
cosinesim.sort(key=lambda x:x[1],reverse =True)
simusers=cosinesim[0:6]
simusers
```

```python
'''Use the “movies” dataset to find out the names of movies, user 1 and user 338 have 
watched in common and how they have rated each one of them'''
movies = pd.read_csv('movies.csv')
movies
comm = pd.DataFrame(pd.merge(df, movies, on="movieId"))
u1 = pd.DataFrame(comm[comm["userId"] == 1][["title","rating"]])
u338 = pd.DataFrame(comm[comm["userId"] == 338][["title","rating"]])
df_2=pd.DataFrame(pd.merge(u1,u338))
df_2.head()
```

```python
''' Use the movies dataset to find out the common movie names between user 2 and user 
338 with least rating of 4.0'''
u2=pd.DataFrame(comm[comm["userId"] == 2][["title","rating"]])
u338=pd.DataFrame(comm[comm["userId"] == 338][["title","rating"]])
df_3=pd.DataFrame(pd.merge(u2,u338))
df_4=df_3[df_3['rating']>=4.0]
df_4.head()
```
